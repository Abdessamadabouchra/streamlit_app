```python
from selenium import webdriver
from selenium.webdriver.common.by import By
 
driver= webdriver.Chrome()
driver.get("https://www.amazon.com/")

try:
    driver.find_element(By.XPATH, '//*[@id="twotabsearchtextbox"]').send_keys("notebook")
    driver.find_element(By.XPATH, '//*[@id="nav-search-submit-button"]').click()
    products= driver.find_elements(By.XPATH, '//*[@id="search"]/div[1]/div[1]/div/span[1]/div[1]/div[8]/div/div/div/div/div[2]/div[1]/h2/a/span')
    prix= driver.find_elements(By.XPATH, '//*[@id="search"]/div[1]/div[1]/div/span[1]/div[1]/div[8]/div/div/div/div/div[2]/div[3]/div/a/span[1]/span[2]/span[2]')
   
    for produit in products:
        print("produit est ",produit.text)
    for prix in prix:
        print("prix est ",prix.text)
except:
    print ("error my friend !")
```
